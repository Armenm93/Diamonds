```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL de la page Wikipedia
url = "https://en.wikipedia.org/wiki/List_of_diamonds#List_of_diamonds"

# Faire la requête HTTP
response = requests.get(url)

# Vérifier si la requête a réussi
if response.status_code == 200:
    # Parser le contenu HTML
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Trouver le tableau sur la page
    table = soup.find('table', {'class': 'wikitable'})

    # Extraire les données du tableau
    rows = table.find_all('tr')
    
    # Créer une liste pour stocker les données
    data = []

    # Parcourir les lignes du tableau
    for row in rows:
        # Trouver toutes les cellules dans la ligne
        cells = row.find_all(['th', 'td'])
        # Extraire le texte de chaque cellule
        row_data = [cell.get_text(strip=True) for cell in cells]
        # Ignorer la ligne si elle est vide
        if row_data:
            # Ignorer la colonne d'image
            if row_data[0] != "Image":
                data.append(row_data)

    # Convertir les données en DataFrame en utilisant pandas
    df_countryorigin= pd.DataFrame(data[1:], columns=data[0])

df_countryorigin
```
