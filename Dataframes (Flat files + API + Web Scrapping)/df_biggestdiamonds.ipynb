```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Récupération du contenu de la page
url = "https://en.wikipedia.org/wiki/List_of_largest_rough_diamonds"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

# Recherche du tableau contenant les données
table = soup.find("table", {"class": "wikitable sortable"})

# Extraction des données du tableau dans une liste de listes
data = []
rows = table.find_all("tr")
for row in rows:
    cols = row.find_all(["th", "td"])
    cols = [col.text.strip() for col in cols]
    data.append(cols)

# Création du DataFrame à partir des données
df_biggestdiamonds = pd.DataFrame(data[1:], columns=data[0])

# Affichage du DataFrame
df_biggestdiamonds
```

```python
import re 

def clean_references(x):
    return re.sub("\[[0-9]+\]","", x).strip()

#df_countryorigin["Name"] = df_countryorigin["Name"].a
df_biggestdiamonds = df_biggestdiamonds.applymap(clean_references)
```

```python
df_biggestdiamonds.replace('', 'Unknown', inplace=True)
```

```python
df_biggestdiamonds.to_csv('df_biggestdiamonds.csv')
```
